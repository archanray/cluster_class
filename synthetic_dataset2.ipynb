```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# one line segment

x = np.arange(-10, -5, )
```

```python
lineseg0 = lambda a : 5*a + 20
lineseg1 = lambda a : 2*a + 5
lineseg2 = lambda a : (-0.5*a) + 5
lineseg3 = lambda a : 5*a - 22.5

line_segments = [lineseg0, lineseg1, lineseg2, lineseg3]

x_range = [(-10, -5), (-5, 0), (0,5), (5,10)]

x_values = []
y_line_values = []
y_values = []


for i in range(4):
    
    x = list(np.random.uniform(low=x_range[i][0], high=x_range[i][1], size=(500,)))
    y_line = [line_segments[i](val) for val in x]
    y = [line_segments[i](val) + np.random.normal(0,2.5) for val in x] 
    
    x_values.extend(x)
    y_line_values.extend(y_line)
    y_values.extend(y)
    
```
